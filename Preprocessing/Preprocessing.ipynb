# IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
```

# LOAD DATA

```python
df = pd.read_csv("dataset_after_EDA.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

# BERSIHKAN NAMA KOLOM & ISI

## Hilangkan spasi tersembunyi

```python
df.columns = df.columns.str.strip()
```

## Bersihkan spasi pada data kategorikal

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
```

# UBAH NAMA KOLOM → BAHASA INDONESIA

```python
rename_columns = {
    'loan_id': 'id_pinjaman',
    'no_of_dependents': 'jumlah_tanggungan',
    'education': 'pendidikan',
    'self_employed': 'wirausaha',
    'income_annum': 'pendapatan_tahunan',
    'loan_amount': 'jumlah_pinjaman',
    'loan_term': 'jangka_waktu_pinjaman',
    'cibil_score': 'skor_kredit',
    'residential_assets_value': 'aset_rumah',
    'commercial_assets_value': 'aset_usaha',
    'luxury_assets_value': 'aset_mewah',
    'bank_asset_value': 'aset_bank',
    'loan_status': 'status_pinjaman'
}
```

```python
df.rename(columns=rename_columns, inplace=True)
```

```python
df.columns
```

# DROP KOLOM ID KARENA TIDAK DIGUNAKAN

```python
df = df.drop(columns=['id_pinjaman'])
```

```python
df.head()
```

# MEYIMPAN DATASET VERSI BAHASA INDONESIA (UNTUK JAGA JAGA)

```python
df.to_csv("dataset_preprocessing_bahasa_indonesia.csv", index=False)
```

# MENGECEK KEMBALI MISSING VALUE 

```python
df.isnull().sum()
```

# Memisahkan FITUR (X) DAN TARGET (y)

```python
X = df.drop(columns=['status_pinjaman'])
y = df['status_pinjaman']
```

```python
X.head()
```

```python
y.value_counts()
```

# Mengidentifikasi TIPE DATA

```python
kolom_kategorikal = X.select_dtypes(include='object').columns.tolist()
kolom_numerik = X.select_dtypes(exclude='object').columns.tolist()

print("Kolom kategorikal:", kolom_kategorikal)
print("Kolom numerik:", kolom_numerik)
```

# ENCODING DATA KATEGORIKAL

```python
X = pd.get_dummies(X, columns=kolom_kategorikal, drop_first=True)
```

# SCALING DATA NUMERIK

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X[kolom_numerik] = scaler.fit_transform(X[kolom_numerik])
```

# SIMPAN DATASET SIAP MODEL

```python
X.to_csv("X_preprocessing_ready.csv", index=False)
y.to_csv("y_target.csv", index=False)
```
