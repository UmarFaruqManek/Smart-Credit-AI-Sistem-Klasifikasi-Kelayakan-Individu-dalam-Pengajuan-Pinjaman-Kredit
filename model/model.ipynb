# IMPORT LIBRARY MODELING

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# LOAD DATASET

```python
X = pd.read_csv("X_preprocessing_ready.csv")
y = pd.read_csv("y_target.csv")
```

```python
y = y.values.ravel()
```

# ENCODING TARGET

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Mapping label:")
for i, label in enumerate(le.classes_):
    print(f"{label} → {i}")
```

```python
np.unique(y_encoded)
```

# Modelling 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from xgboost import XGBClassifier
```

## Split Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,   # target yang SUDAH encoding
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### LOGISTIC REGRESSION

```python
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

```python
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42
)

rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_lr, target_names=['Approved', 'Rejected']))
```

### Random Forest

```python
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42
)

rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_rf, target_names=['Approved', 'Rejected']))
```

### XGBoost

```python
xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='logloss',
    random_state=42
)

xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_xgb, target_names=['Approved', 'Rejected']))
```

### EVALUASI KETIGA MODEL 

```python
print("Logistic Regression Accuracy :", accuracy_score(y_test, y_pred_lr))
print("Random Forest Accuracy       :", accuracy_score(y_test, y_pred_rf))
print("XGBoost Accuracy             :", accuracy_score(y_test, y_pred_xgb))
```

```python
import joblib

# Simpan ketiga model
joblib.dump(lr, 'logistic_regression_model.pkl')
joblib.dump(rf, 'random_forest_model.pkl')
joblib.dump(xgb, 'xgboost_model.pkl')

print("✓ Model Logistic Regression tersimpan sebagai: logistic_regression_model.pkl")
print("✓ Model Random Forest tersimpan sebagai: random_forest_model.pkl")
print("✓ Model XGBoost tersimpan sebagai: xgboost_model.pkl")
```
